```python
import os
import docx
import json

def convert_docx_to_json(docx_file_path):
    # Word belgesini adından unite adını al
    unite_name = os.path.splitext(os.path.basename(docx_file_path))[0]

    # Word belgesini aç
    doc = docx.Document(docx_file_path)

    # Diyalogları içeren liste
    dialogs = []

    # Geçerli diyalog
    current_dialog = {"kelime": "", "anlami": ""}

    for para in doc.paragraphs:
        text = para.text.strip()

        # Numaralı başlıkları kontrol et
        if text.endswith("-"):
            if current_dialog["kelime"] and current_dialog["anlami"]:
                dialogs.append(current_dialog)
                current_dialog = {"kelime": "", "anlami": ""}

        # Türkçe diyalogları ekleyin
        if text.startswith("T: "):
            current_dialog["kelime"] += text.replace("T: ", "") + " "

        # İngilizce diyalogları ekleyin
        elif text.startswith("İ: "):
            current_dialog["anlami"] += text.replace("İ: ", "") + " "

    # Son diyalogu ekle
    if current_dialog["kelime"] and current_dialog["anlami"]:
        dialogs.append(current_dialog)

    # JSON formatına çevir
    json_data = {unite_name: dialogs}

    # JSON verisini döndür
    return json_data

def merge_jsons(json_files):
    merged_data = {}
    for file_name, json_data in json_files.items():
        merged_data.update(json_data)
    return merged_data

# Word dosyalarının bulunduğu klasör
folder_path = "C:/Users/numan/OneDrive/Desktop/ah"

# Tüm Word dosyalarını al
word_files = [os.path.join(folder_path, file) for file in os.listdir(folder_path) if file.endswith(".docx")]

# Her bir Word dosyasını JSON formatına çevir
json_data_list = {os.path.splitext(os.path.basename(file))[0]: convert_docx_to_json(file) for file in word_files}

# Tüm JSON verilerini birleştir
merged_json_data = merge_jsons(json_data_list)

# Birleştirilmiş JSON verisini dosyaya yaz
output_file_path = "merged_data.json"
with open(output_file_path, "w", encoding="utf-8") as output_file:
    json.dump(merged_json_data, output_file, ensure_ascii=False, indent=4)

print(f"Tüm Word dosyalarının birleştirilmiş JSON verisi '{output_file_path}' dosyasına kaydedildi.")
```
